```python
import plotly.express as px

df["OrderDate"] = pd.to_datetime(df["OrderDate"])

monthly_sales = df.groupby(df["OrderDate"].dt.to_period("M"))["Sales"].sum()

monthly_sales.index = monthly_sales.index.to_timestamp()

fig = px.line(
    x=monthly_sales.index,
    y=monthly_sales.values,
    labels={"x": "Month", "y": "Sales"},
    title="Monthly Sales Trend"
)

fig.show()
```

```python
region_analysis = df.groupby("Region")[["Sales", "Profit"]].sum().reset_index()

region_analysis
```

```python
fig = px.bar(
    region_analysis,
    x="Region",
    y=["Sales", "Profit"],
    barmode="group",
    title="Sales and Profit by Region"
)

fig.show()
```

```python
category_analysis = df.groupby("Category")["Sales"].sum().reset_index()

category_analysis
```

```python
fig = px.pie(
    category_analysis,
    names="Category",
    values="Sales",
    title="Sales Distribution by Category"
)

fig.show()
```

```python
top_customers = (
    df.groupby("CustomerName")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_customers
```

```python
fig = px.bar(
    top_customers,
    x="CustomerName",
    y="Sales",
    title="Top 10 Customers"
)

fig.show()
```

```python
yearly_sales = df.groupby("Year")["Sales"].sum().reset_index()

yearly_sales
```

```python
fig = px.line(
    yearly_sales,
    x="Year",
    y="Sales",
    markers=True,
    title="Year-over-Year Sales Growth"
)

fig.show()
```

```python
subcategory_analysis = (
    df.groupby("SubCategory")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

fig = px.bar(
    subcategory_analysis,
    x="SubCategory",
    y="Sales",
    title="Sales by SubCategory"
)

fig.show()
```
